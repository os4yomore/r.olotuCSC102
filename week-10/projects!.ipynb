```python
#Project 1

class oranges():
    def __init__(self, quantity):
        self.quantity = quantity
        
    def thanks(self):
        print("Thank you for your patronage!")
        
    def calculate(self):
        if self.quantity > 30:
            print("Unfortunately, your demand exceeds our stock profile")
            x = input("Perhaps you would like to select a smaller quantity. If yes, type yes, if not, type no. ")
            if x.lower() == "no":
                exit()
            elif x.lower() == "yes":
                new = int(input("What is your new needed quantity?  "))
                price = new * 200
                print("Your bill is N", price)
        elif self.quantity <= 30:
            price = self.quantity * 200
            print("Your bill is N", price)
            
print("Welcome to Ruth's Orange Shop")
print("Cost = N200 per orange")
number = int(input("How many oranges would you like to buy?  "))
object = oranges(number)
object.calculate()
object.thanks()
            
            
        
```

```python
#project 2
class major():
    def __init__(self, result):
        self.result = result
        
    def finalanswer(self):
        print("Your final answer is: ", self.result)
    
class trap(major):
    def __init__(self, height, b1, b2):
        self.height = height
        self.b1 = b1
        self.b2 = b2
        
    def calculate(self):
        self.result = (self.height / 2) * (self.b1 + self.b2)

class rhom(major):
    def __init__(self, d1, d2):
        self.d1 = d1
        self.d2 = d2
        
    def calculate(self):
        self.result = 0.5 * self.d1 * self.d2

class para(major):
    def __init__(self, b1, alt):
        self.b1 = b1
        self.alt = alt
        
    def calculate(self):
        self.result = self.b1 * self.alt 

class cube(major):
    def __init__(self, length):
        self.length = length
        
    def calculate(self):
        self.result = 6 * self.length * self.length
        
class volume(major):
    def __init__(self, radius, height):
        self.radius = radius
        self.height = height
        
    def calculate(self):
        self.result = 3.1415 * self.radius * self.radius * self.height
        
print("Welcome to Ruth's Calculator!")
print("Click 1 for the area of a trapezium")
print("Click 2 for the area of rhombus")
print("Click 3 for the area of parallelogram")
print("Click 4 for the surface area of a cube")
print("Click 5 for the volume of a cylinder")

option = int(input("What would you like to calculate?   "))

if option == 1:
    a = int(input("Input height:  "))
    b = int(input("Input base 1:  "))
    c = int(input("Input base 2:  "))
    object = trap(a, b, c)
    object.calculate()
    object.finalanswer()
    
if option == 2:
    a = int(input("Input diagonal 1:  "))
    b = int(input("Input diagonal 2:  "))
    object = rhom(a, b)
    object.calculate()
    object.finalanswer()
    
if option == 3:
    a = int(input("Input base:  "))
    b = int(input("Input altitude:  "))
    object = para(a, b)
    object.calculate()
    object.finalanswer()
    
if option == 4:
    a = int(input("Input length of side:  "))
    object = cube(a)
    object.calculate()
    object.finalanswer()
    
if option == 5:
    a = int(input("Input radius:  "))
    b = int(input("Input height:  "))
    object = volume(a, b)
    object.calculate()
    object.finalanswer()
        
        
        
        
        
        
        
        
        
        
        
        
        
        
        
        
        
        
        
        
        
        
        
        
```
