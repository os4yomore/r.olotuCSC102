# CONNECTING PYTHON WITH POSTGRES DATABASE 

### 1. CREATE A DATABSE student_profile

```python
import psycopg2

# Establish a connection to the PostgreSQL server
conn = psycopg2.connect(
    host="localhost",
    port="5432",
    user="postgres",
)

# Create a new database
new_db_name = "student_profile"
conn.autocommit = True
cursor = conn.cursor()
cursor.execute(f"CREATE DATABASE {new_db_name};")

cursor.close()


# Close the connection
conn.close()
```

### 2. CREATE A TABLE student_info

```python
# Establish a connection to the PostgreSQL server
conn = psycopg2.connect(
    host="localhost",
    port="5432",
    database="student_profile",
    user="postgres",
    
)
# Create a cursor object
cursor = conn.cursor()

# Define the CREATE TABLE statement
create_table_query = '''
    CREATE TABLE student_info (
        name varchar,
        email varchar,
        mat_no varchar  )
'''

# Execute the CREATE TABLE statement
cursor.execute(create_table_query)


# Commit the changes
conn.commit()

# Close the cursor and connection
cursor.close()
conn.close()
```

### 4. SELECT ALL FROM student_info

```python
conn = psycopg2.connect(
    host="localhost",
    port="5432",
    database="comp_sci_dpt",
    user="postgres",
    
)
# Create a cursor object

cursor = conn.cursor()

# Define the CREATE TABLE statement
create = '''
    CREATE TABLE csc102_class_list (
        FIRST_NAME varchar,
        MIDDLE_NAME varchar,
         SURNAME  varchar)
        
'''
cursor.execute(create)
conn.commit()
cursor.close()
conn.close()
```

```python
import pandas as pd

conn = psycopg2.connect(
    host="localhost",
    port="5432",
    database="comp_sci_dpt",
    user="postgres"
)
cursor = conn.cursor()

data = pd.read_csv("csc102_class_list.csv")
data2 = data[['FIRST NAME', 'MIDDLE NAME', 'SURNAME']]

cursor.execute('''
    CREATE TABLE IF NOT EXISTS csc102_class_list (
        "FIRST NAME" VARCHAR,
        "MIDDLE NAME" VARCHAR,
        "SURNAME" VARCHAR
    )
''')


# Insert the data into the table
for row in data2.itertuples(index=False):
    cursor.execute(
        'INSERT INTO csc102_class_list (FIRST_NAME, MIDDLE_NAME, SURNAME) VALUES (%s, %s, %s)',
        row
    )

# Commit the changes
conn.commit()

# Close the cursor and connection
cursor.close()
conn.close()
```
